# 2. Set upper lower level, get lists

```python
L1=315
L2=837

R1=320
R2=842

x = np.arange(L1, L2+1, 6)
list_1 = x.tolist()

y = np.arange(R1, R2+1, 6)
list_2 = y.tolist()
```

```python
print(list_1)
print(list_2)
```

# 3. Get left and right list put to dataframe

```python
df2 = pd.DataFrame(list(zip(list_1, list_2)), columns =['Left', 'Right'])
df2['diff']=df2['Right']-df2['Left']
df2.head(10)
df2.info()
```

# 4. Combine Left and right list and sort

```python
band=list_1+list_2
band.sort()
print(band)
```

# 5. Create sample data for score

```python
# set a random seed to reproduce
np.random.seed(123)
# create 10 random integers
x = np.random.randint(low=315, high=840, size=200)
x
```

```python
df = pd.DataFrame({"score":x})
df.head(5)
#df.info()
#df.astype(int)
```

```python
for v in df['score']:
    df3 = df2[(df2['Left'] < v) &  (df2['Right'] > v)]
    print (df3)
```

```python
df = df.assign(a=1).merge(df2.assign(a=1), on='a', how='outer')
print (df)
```

```python
df3 = df[(df['Left'] < df['score']) &  (df['Right'] > df['score'])]
#print (df3)
df3.head(5)

df4 = df3.sort_values(by=['score'], ascending=True)
df4.head(25)
```
